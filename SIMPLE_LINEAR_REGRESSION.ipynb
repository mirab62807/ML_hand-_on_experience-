# Importing Libraries

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

# Import data

```python
data = pd.read_csv('/content/Salary_Data.csv')
data.head()
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1].values
```

```python
x
```

#Split data_set

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reggressor = LinearRegression()
reggressor.fit(x_train, y_train)
```

```python
y_predict = reggressor.predict(x_test)
y_predict
```

# train data

```python
plt.scatter(x_train, y_train, color ='red')
plt.plot(x_train, reggressor.predict(x_train))
plt.show()
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_train, reggressor.predict(x_train))
plt.show
```

```python
pre_yr12 = reggressor.predict([[12]])
pre_yr12
```

```python
print(reggressor.coef_)
print(reggressor.intercept_)
```
